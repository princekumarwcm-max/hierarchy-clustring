# hierarchical clustring

```python
# hierarchical clustring:-jisme data points ko unki similarity ke basis par ek tree-like structure (hierarchy) me arrange kiya jata hai.
# Yeh method step-by-step similar data points ko merge (jodta) ya split (alag) karta hai taaki clusters ban sake.

# Dendrogram in hierarchical clustring:-Ye diagram clusters ke merge hone (ya split hone) ka visual representation deta hai.
# agglomerative hierarchical practical
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv(r"C:\Users\Prince Kumar\Downloads\iris_dataset.csv")
```

```python
dataset.head(4)
```

```python
sns.pairplot(data=dataset)
plt.show()
```

```python
# dandrogram banana ka liya scipy library download karni hai ok pip install Scipy
import scipy.cluster.hierarchy as sc
```

```python
x = dataset.drop(['target'], axis=1)

# Create dendrogram
plt.figure(figsize=(10,10))
dendro = sc.dendrogram(sc.linkage(x, method='single', metric='euclidean'))
# plt.title("Dendrogram (Single Linkage)")
# plt.xlabel("Samples")
# plt.ylabel("Euclidean Distance")
plt.savefig("demo.jpg")
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering
```

```python
ac = AgglomerativeClustering(n_clusters=2,linkage='single')
ac.fit_predict(x)
```

```python
ac = AgglomerativeClustering(n_clusters=2,linkage='single')
dataset["Predict"] = ac.fit_predict(x)
```
